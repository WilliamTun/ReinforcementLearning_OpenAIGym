# Temporal Difference Learning

- This is a model free learning algorithm 
- and it can be applied to non-episodic tasks

TD uses Monte Carlo and Dynamic Programming

TDL approximates a current estimate based on previously learned estimate - which is also called bootstrapping. 

- TD tries to predict state values. 
- In Monte Carlo, es estimate value function by simply taking mean return
- TD updates value at previous state - by current tate. - using TD update rule for updating value of state. 


OldStateValue_updated[t-1] = OldStateValue[t-1] + learning_rate(reward + discount_factor(StateValue[1])-OldStateValue[t-1])

^ This equation is the difference between the actual reward and the expected reward... multiplied by learning rate("step size"). This residual error is called the "TD error" which we try to minimise per iteration 

# TD control

A estimate of the value function needs to be optimised in TD predicion which is done in two ways:
    
- Off policy (eg. Q learning)
- On policy (SARSA)

# OFF POLICY TD control - Q learning

- 1. initialise a Q function to arbitary values
- 2. take an action using episolon greedy policy (random --> less random)
- 3. use equation to update previous state 
- 4. repeat steps 2 and 3 until termination

```python
# In the frozen lake examples the Q-table is set up as 
# State          Action       Q-Value 
# location(3,2)   Up           0.1
# location(3,2)   Down         0.5
# location(3,2)   Left         0.3
```

```python
# Start 
# Initalise random Q values + define a terminal state
# LOOP:
#     Start episode
#     Choose action via episolon greedy policy
#     perform action to new state and recieve reward
#     Update Q value of previous state
#     If State = terminal state:
#         break
```

```python
import gym 
import random
from gym import envs
```

```python
envs.registry.all()
```

```python
env = gym.make("Taxi-v2")
```

```python
env.render() # show taxi environment
```

```python
# Initialise parameters
alpha = 0.4 
gamma = 0.999
epsilon = 0.17
```

```python
q = {}
for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        q[(s,a)] = 0.0 
```

```python
q 

# Here is a Q table where every state has 5 actions and initial 0 q values
# (state, action): q value
```

```python
def update_q_table(prev_state, action, reward, nextstate, alpha, gamma):
    # loop through all possible actions , and choose the action and next state based on the largest q value
    qa = max([q[(nextstate, a)] for a in range(env.action_space.n)]) 
    # update q values of previous state/action in q table using formula
    q[(prev_state, action)] += alpha * (reward + gamma * qa - q[(prev_state, action)])
    
    
```

```python
def epsilon_greedy_policy(state, epsilon):
    # if epsilon is greater than the randomly generated number, randomly sample and action
    if random.uniform(0,1) < epsilon:
        return env.action_space.sample()
    else:
    # else, use the action with the highest q value for the given action for the given state
        return max(list(range(env.action_space.n)), key = lambda x: q[(state, x)])
```

```python
for i in range(8000):
    reward_storage = 0 
    prev_state = env.reset()
    while True:
        # in each state, select action with epsilon greedy policy:
        action = epsilon_greedy_policy(state=prev_state, epsilon=epsilon)
        # do action and move to next state and recieve reward
        next_state, reward, done, _ = env.step(action)
        
        # update q values and previous state
        update_q_table(prev_state=prev_state, 
                       action=action, 
                       reward=reward, 
                       nextstate=next_state, 
                       alpha=alpha, 
                       gamma=gamma)
        prev_state = next_state
        
        # store rewards
        reward_storage += reward
        
        if done:
            break
    print("total reward: ", reward_storage)

env.close()
        
```
