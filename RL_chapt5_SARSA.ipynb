# State-Action-Reward-State-Action

- on policy TD control algorithm
- Q learning: State-value pairs 
- SARSA : state action value triplets

SARSA uses an update rule: 
- Q(s,a) = Q(s,a) + alpha(reward + (gamma * Q (s', a')) - Q(s,a))

```python
# In Q learning, when take an action based on the one with the highest q value. 
# The main difference between SARSA and Q value learning...
# In SARSA, we take an action ALSO using an epislon greedy policy. 
```

```python
import gym
import random
env = gym.make('Taxi-v2')
```

```python
# Initialise Q table of all states and all possible actions for each state
Q = {}
for s in range(env.observation_space.n):
    for a in range(env.action_space.n):
        Q[(s,a)] = 0.0
```

```python
def epsilon_greedy(state, epsilon):
    if random.uniform(0,1) < epsilon:
        return(env.action_space.sample())
    else:
        return(max(list(range(env.action_space.n)), key = lambda x: Q[(state,x)]))
```

```python
# Initialise parameters
alpha = 0.85
gamma = 0.9 
epsilon = 0.8 

for i in range(4000):
    # We store cumulative reward of each spisodes in r
    r = 0 
    # Then for every iterations, we initialize the state
    state = env.reset()
    # then pick action using epsilon greedy
    action = epsilon_greedy(state, epsilon)
    
    while True:
        nextstate, reward, done, _ = env.step(action)
        nextaction = epsilon_greedy(nextstate, epsilon)
        Q[(state, action)] += alpha * (reward + gamma * Q[(nextstate, nextaction)] - Q[(state, action)])
        # update state and action with next action and next state
        action = nextaction
        state = nextstate
        r += reward 
        # we will break the loop, if we are at the terminal
        # state of the episode
        if done:
            break
```
